### Regression Losses (Continuous output)
**MSE (Mean Squared Error)**
- L=n1​∑(y−y^​)2
- Penalizes large errors heavily. Used in regression tasks
  
**MAE (Mean Absolute Error)**
- L=n1​∑∣y−y^​∣
- More robust to outlier that MSE

### Classification Losses 
**Binary Classification (2 classes)**
- **Binary Cross-Entropy (BCE)**
- L=−[ylog(y^​)+(1−y)log(1−y^​)]
- Used when output is sigmoid. Eg Spam yes/no

**Multi-class classification (>2 classes)**
- **Cross Entropy loss**
- L=−∑yi​log(y^​i​)
- Used with softmax output

***Don't use softmax at the end of the NN if we are using CrossEntropy, The CE automatically adds the softmax, So no need***

***For Binary CrossEntropy we need to add the fixed size output in the NN and we need to add sigmoid function at the end***

#### BCEWithLogits
***Special case, If the final layer have only one node, we can use nn.BCEWithLogitsLoss() so we dont need to mention sigmoid also in the end of NN***

```python
# Cross Entropy Loss (Multi class Classification)


import torch
import torch.nn as nn

# Suppose we have 3 classes and 4 samples
logits = torch.tensor([[2.0, 1.0, 0.1],
                       [0.5, 2.5, 0.3],
                       [1.0, 2.0, 0.1],
                       [2.2, 0.1, 1.5]])   # [batch, num_classes]

labels = torch.tensor([0, 1, 1, 2])        # ground truth class indices

criterion = nn.CrossEntropyLoss()
loss = criterion(logits, labels)
print("CrossEntropy Loss:", loss.item())
```

```python
# BCELoss (Binary classification, with sigmoid)

import torch
import torch.nn as nn

# 4 samples, single output per sample
logits = torch.tensor([0.8, -1.2, 2.5, 0.0])    # raw values
labels = torch.tensor([1, 0, 1, 0], dtype=torch.float32)  # binary labels

# Apply sigmoid manually
probs = torch.sigmoid(logits)

criterion = nn.BCELoss()
loss = criterion(probs, labels)
print("BCE Loss:", loss.item())
```

```python
# BCEWithLogitsLoss (Binary classification, stable way)

import torch
import torch.nn as nn

logits = torch.tensor([0.8, -1.2, 2.5, 0.0])   # raw values
labels = torch.tensor([1, 0, 1, 0], dtype=torch.float32)

criterion = nn.BCEWithLogitsLoss()
loss = criterion(logits, labels)   # sigmoid applied inside
print("BCEWithLogits Loss:", loss.item())
```
